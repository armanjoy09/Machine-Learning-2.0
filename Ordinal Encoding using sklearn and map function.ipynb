```python
import pandas as pd
df=pd.DataFrame({"size":["m","s","l","xl","xxl","s","m","l","xl","xxl","l","s"]})
```

```python
df.head()
```

```python
ord_data=[['s','m','l','xl','xxl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
```

```python
oee=oe.fit_transform(df[["size"]])
```

```python
df["oe_size"]=oee
```

```python
df
```

# map function

```python
or_data1={'s':1,'m':2,'l':3,'xl':4,'xxl':5}
```

```python
df["oe_size_map"]=df["size"].map(or_data1)
```
