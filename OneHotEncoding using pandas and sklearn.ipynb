```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"D:\ML Code\loan.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# filling gender columns with most frequently used categories
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True)
```

```python
en_dataset=dataset[["Gender","Married"]]
```

```python
en_dataset
```

```python
pd.get_dummies(en_dataset)
```

```python
pd.get_dummies(en_dataset).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ar=ohe.fit_transform(en_dataset).toarray()
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Married_No","Married_Yes"])
```

```python
# drop function that delete the first culumns of the generated columns using onehot encoder
# like Gender_female and Married_No
ohe=OneHotEncoder(drop="first")
```

```python
arr=ohe.fit_transform(en_dataset).toarray()
```

```python
pd.DataFrame(arr,columns=["Gender_Male","Married_Yes"])
```
